### 2_2- Indexing in pandas

Les plus utilisés surtout dans les operations avancées
- iloc (sélection basée sur l'index : sélection des données en fonction de leur position numérique dans les données)
- loc

#### 2_2_1- Index-Based Selection (iloc)

```python
# affichage des caracteristiques de toute la première ligne (ou enregistrement)
reviews.iloc[0]
```

```python
# affichage des enregistrement avec leurs 1ers caracteristiques (ou propriété)
reviews.iloc[:, 0]
```

```python
# Selection de la 1ère, 2ème et 3ème ligne de la 1ère colonne
reviews.iloc[:3, 0]
```

```python
# 2ème et 3ème enregistrement
reviews.iloc[1:3, 0]
```

```python
# Une liste peut également être passé dans le []
reviews.iloc[[0, 1, 60, 777], 0]
```

```python
# Utilisation de nombres négatifs pour la selection
reviews.iloc[-5:]
#C'est tout comme avec les listes en python
```

#### 2_2_2- Label-Based Selection (loc)
c'est la valeur de l'index de données (donc pas l'indice elle même), pas sa position, qui compte. il est conceptuellement plus simple que iloc. Il traite le Dataset comme un gros matrice (une liste de liste)

```python
reviews.loc[0, 'country']
```

```python
reviews.loc[:, ['taster_name', 'taster_twitter_handle', 'points']]
```

### loc Vs iloc
Ceci est particulièrement déroutant lorsque l'index DataFrame est une simple liste numérique, par ex. 0,...,1000. Dans ce cas, df.iloc[0:1000] renverra 1000 entrées, tandis que df.loc[0:1000] en renverra 1001 ! Pour obtenir 1000 éléments en utilisant loc, vous devrez descendre un plus bas et demander df.loc[0:999]

### Manipulation des indexes
loc derive sa puissance de iloc.
- Les indices que nous manipulons sont immuables, mais la methode set_index() peut faire le boulot

```python
# utiliser la colonne "title" comme indexe au lieu des chiffres (0, 1, 2, ..)
reviews.set_index("title")
```

#### - Conditional Selection

```python
reviews.country == 'Italy'
```

```python
# Affichage desenregistrement ayant pour ville "Italy"
reviews.loc[reviews.country == 'Italy']
```

```python
# Affichage des vins d'Italy accumulant au moins 90 points
reviews.loc[(reviews.country == 'Italy') & (reviews.points >= 90)]
```

```python
# vin fabriqué en Italie ou classé au-dessus de la moyenne
reviews.loc[(reviews.country == 'Italy') | (reviews.points >= 90)]
```

Deux (2) Selecteurs conditionels intégrés dans pandas:
- isin
- isnull et son compagnon "notnull"

```python
reviews.loc[reviews.country.isin(['Italy', 'France'])]
```

```python
reviews.loc[reviews.price.notnull()]
```

### 3_2-Maps
Il existe deux méthodes de mappage que vous utiliserez souvent.
- map()
- apply()

```python
review_points_mean = reviews.points.mean()
reviews.points.map(lambda p: p - review_points_mean)
```

```python
def remean_points(row):
    row.points = row.points - review_points_mean
    return row

reviews.apply(remean_points, axis='columns')
```

```python
#Rien n'a changé sur la 1ère ligne
reviews.head(1)
```

```python
review_points_mean = reviews.points.mean()
reviews.points - review_points_mean
```

```python
"""What is the difference between map and apply?"""

import pandas as pd


df = pd.DataFrame([(1, 2, 3), (5, 6, 7), (9, 0, 1), (3, 4, 5)],
                  columns=list('abc'),
                  index=['India', 'France', 'England', 'Germany'])
df['b'] = df['a']
df['c'] = df['a']
print(df)
# Is there a difference between .map and .apply?
df['b'] = df['b'].map(lambda x: x**2)
print(df)

df['c'] = df['c'].apply(lambda x: x**2)
print(df)
```

## IV-Grouping and Sorting

```python
import pandas as pd
reviews = pd.read_csv(r"C:\Users\Oussou\Documents\Rachid Ousmane\cours auto\ODC_AI\Veille3\wine.csv", index_col=0)
pd.set_option('display.max_rows', 5)
```

```python
reviews.groupby('points').points.count()
```

```python
reviews.groupby('points').price.min()
```

```python
reviews.groupby('winery').apply(lambda df: df.title.iloc[0])
```

```python
reviews.groupby(['country', 'province']).apply(lambda df: df.loc[df.points.idxmax()])
```

```python
reviews.groupby(['country']).price.agg([len, min, max])
```

```python
countries_reviewed = reviews.groupby(['country', 'province']).description.agg([len])
countries_reviewed
```

```python
mi = countries_reviewed.index
type(mi)
```

```python
countries_reviewed.reset_index()
```

### Sorting

```python
countries_reviewed = countries_reviewed.reset_index()
countries_reviewed.sort_values(by='len')
```

```python
countries_reviewed.sort_values(by='len', ascending=False)
```

```python
countries_reviewed.sort_index()
```
